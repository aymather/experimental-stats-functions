```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split as tts
import statsmodels.api as sm
import pandas as pd
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
data = pd.read_csv('jaybob.csv')
data.head()
```

```python
def logit(x, y):
    
    # Split into training and testing data
    x_train, x_test, y_train, y_test = tts(x, y, train_size=0.7, random_state=1)
    
    # Build our model
    model = LogisticRegression().fit(x_train, y_train)

    # Make predictions and get probabilities
    y_predict = model.predict(x_test)
    y_predict_probabilities = model.predict_proba(x_test)
    ypp = y_predict_probabilities[:, 1]
    
    # Build a plot to demonstrate our predictions
    plt.scatter(x_test, y_predict, color='blue', label='Test Predictions')
    plt.scatter(x_test, ypp, color='red', label='Probabilities')
    plt.legend()
    plt.show()
    
    plt.scatter(x_test, y_test, color='green', label='Test True Values')
    plt.legend()
    plt.show()
```

```python
# Independent Variables
x = data['Price'].values.reshape(-1, 1)

# Dependent Variable
y = data['Sold?']

logit(x, y)
```

```python
# Independent Variable
x = data['Odometer'].values.reshape(-1, 1)

logit(x, y)
```

```python
x = data['Age'].values.reshape(-1, 1)
logit(x, y)
```
